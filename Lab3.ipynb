```python
"""Importación los módulos necesarios para el desarrollo un modelo de regresión lineal."""
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def importdataset():
    """Función para importar la dataset"""
    df = pd.read_csv("winequality.csv")
    print("Dataset importado: ", df.head(10))
    return df
```

```python

def valorvariable(df):
    """Asignación de varaibles x e y"""
    # Dar valor a las variables
    x = df['alcohol']
    y = df['quality']
    return x, y
```

```python
#Función para el cálculo de la regresión lineal
def calcreg(x, y):
    # Calcule la regresión lineal de mínimos cuadrados para dos conjuntos de datos.
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    return slope, intercept, r, p, std_err
```

```python
# Utilizar la función auxiliar que usa la pendiente e intersección 
def fitted_value(x, slope, intercept):
    return slope * x + intercept
```

```python
#Función para graficar el resultado
def grafica(x, y, fitted_value):
    # Calcular los valores ajustados para el eje ‘y’ usando una función auxiliar.
    new_values = list(fitted_value)
    # Visualización del modelo.
    plt.plot(x, y, 'o', label='original data')
    # Draw the line or linear regression
    plt.plot(x, new_values, 'r', label='fitted line')
    plt.legend()
```

```python
#Fundión principal
def main():
    data = importdataset()
    x, y = valorvariable(data)
    slope, intercept, r, p, std_err = calcreg(x, y)
    fifvalue = fitted_value(x, slope, intercept)
    grafica(x, y, fifvalue)
```

```python
#Ejecución del código principal
main()
```
